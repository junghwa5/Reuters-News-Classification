```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import reuters
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from math import log

#from tensorflow.keras.preprocessing.sequence import pad_sequences
#from tensorflow.keras.models import Sequential
#from tensorflow.keras.layers import Dense, Embedding, Flatten

from sklearn.feature_extraction.text import TfidfVectorizer
import tensorflow as tf

#from sklearn.feature_extraction.text import CountVectorizer
#from sklearn.feature_extraction.text import TfidfTransformer
#from sklearn.linear_model import SGDClassifier
#from sklearn.pipeline import Pipeline

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=None, test_split=0.2)
```

```python
x_train_list = np.concatenate(x_train.tolist()).tolist()
x_train_list2 = list(set(x_train_list))

word_len = len(x_train_list2) # 전체 문서에서의 단어의 종류
#x_train_list2
```

TF : 1개 문서 안에서 특정 단어의 등장빈도<br>
DF : 특정 단어가 나타나는 문서의 수<br>
IDF : ln((1+n)/(1+df)) + 1

```python
N = len(x_train)

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in x_train:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)*idf(t)
```

```python
result = []
for i in range(N):
    result.append([0 for i in range(word_len)])
    d = x_train[i]
    for j in range(len(d)):
        t = x_train_list2[j]        
        result[i][t-1] = tf(t, d)

tf_ = pd.DataFrame(result, columns = x_train_list2)
tf_
```

```python
result2 = []
for j in range(len(x_train_list2)):
    t = x_train_list2[j]
    result2.append(idf(t))

idf_ = pd.DataFrame(result2, index = x_train_list2, columns = ["IDF"])
idf_
```

```python
# 코드 처리 속도로 인한 오류
result3 = []
for i in range(N):
    result3.append([])
    d = x_train[i]
    for j in range(len(x_train_list2)):
        t = x_train_list2[j]

        result3[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result3, columns = x_train_list2)
tfidf_
```
